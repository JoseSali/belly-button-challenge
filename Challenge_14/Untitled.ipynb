```python
import pandas as pd

def read_excel_header(file_path):
    try:
        # Read the Excel file, specifying the header row
        df = pd.read_excel(file_path, header=0)
        
        # Get the header of the DataFrame
        header = df.columns.tolist()
        
        # Print the header
        print("Header of the Excel file:")
        print(header)
        
        return header
        
    except Exception as e:
        print("Error:", e)

# Provide the path to your Excel file
file_path = "21zp01al (2).xlsx"

# Call the function to read the Excel file's header
header = read_excel_header(file_path)

```
